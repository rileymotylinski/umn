# Recursion
- Anything you can do with recursion, you can do with loops
- what is the smaller version of this problem?

## Laws of Recrusion
1. Must Have a base case
- smallest "solution" to the problem
2. Must reduce input size
- __move__ towards base case
    - e.g. reducing n by one in a recrusive factorial function
## Runtime Complexity
- THe runtime complexity T(N) of a recursive function will have function T on both sides of the equation
1. Single Recursive call T(N-1)
2. One multiplication of the result O(1)
e.g,for factorial, the recurrence relation is. Call + calculation
T(N) = T(N-1) + 1

### Function Stacks
- holds function name, argumentsm etc.

### Recurrence Relation
- T on both sides
T(N) = T(N/2) + 1
- evaluate at all cases

#### Bottom Up
T(0) = 1
T(1) = 1
T(2) = T(N/2) + 1 = T(1) + 1 = 2
T(4) = T(4/2) + 1 = T(2) + 1 = 3
T(8) = 3

#### Top Down
T(N) = T(n/2) = 1
T(n/2) = T(n/4) + 1

1,2,3,4 => Log n
