# Data Structures
### ADT - Abstract Data Type
- not defined by the structure, but rather the operations the user wants to preform
- usually abstracted in your chosen programming language
#### Common ADTs
1. List
2. Dynamic Array
3. Stack - can only insert/read from top of stack
4. Queue - first in, first out
5. Dequeue - inserted, removed from front and back ("double queue")
6. Bag - unsorted list
7. Set
8. Priority Queue - members have priority; higher priority -> moved to front of line
9. Dictionary/Map
